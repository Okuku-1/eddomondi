Exponential Distribution

```python
import numpy as np

def exp_pdf(mu, x):
    delay_rate = 1/mu
    return delay_rate * np.exp(-delay_rate * x)

def exp_cdf(mu, x):
    delay_rate = 1/mu
    return 1 - np.exp(-delay_rate * x)

print("Point probability for exactly 3 minutes: {:.4f}%".format(exp_pdf(4, 3) * 100))
print("Cumulative probability of 3 minutes or less: {:.4f}%".format(exp_cdf(4, 3) * 100))
```

Question 4
The average interval that calls come in at a call center is 8 seconds. Plot the probability density function for a call happening at each second between 0 and 30 seconds (you can look at intervals of 0.5 seconds only.

What is the probability that the next call will happen in 7 seconds?

```python
import matplotlib.pyplot as plt
#given parameters
mu = 8
x_values = np.arange(0, 30.5, 0.5)

#pdf of values
pdf_values = [exp_pdf(mu, x) for x in x_values]

#plotting the PDF
plt.plot(x_values, pdf_values, label="Exponential PDF", color="blue")
plt.xlabel("Time(seconds)")
plt.ylabel('Probability Density')
plt.title('Exponential Distribution: Call Arrival Times')
plt.grid(True)
plt.legend()
plt.show()

# Calculate the probability that the next call will happen at exactly 7 seconds
prob_7_seconds = exp_pdf(mu, 7)
print(prob_7_seconds)  # This will give the PDF value for exactly 7 seconds
```

Question 5
The average earthquake in a given region happens every 7 weeks. What is probability that the next earthquake happens between 5 and 8 weeks from now?

Hint: This has both an upper and lower bound. You'll need to do some arithmetic to solve this one.

```python
lower_bound = exp_cdf(7, 5)
upper_bound = exp_cdf(7, 8)

print("Probability of earthquake before 5 weeks: {}%".format(lower_bound * 100))
print("Probability of earthquake before 8 weeks: {}%".format(upper_bound * 100))
print("Probability of earthquake between 5 - 8 weeks: {}%".format((upper_bound - lower_bound) * 100))
```

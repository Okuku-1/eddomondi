Skewness And Kurtosis

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import kurtosis, skew
```

```python
#setting seed for consistent result
np.random.seed(42)

x_random = np.random.normal(0, 2, 10000)

plt.hist(x_random);

x_skewned = skew(x_random)
x_kurtosis = kurtosis(x_random)

print(f"skewness: {x_skewned}")
print(f"kurtosis: {x_kurtosis}")
```

```python
x = np.linspace( -5, 5, 10000 )
y = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x)**2  )  # normal distribution

plt.hist(y)

y_skewned = skew(y)
y_kurtosis = kurtosis(y)

print(f"skewness: {y_skewned}")
print(f"kurtosis: {y_kurtosis}")
```

Central Limit Theorem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
import scipy.stats as st
np.random.seed(0)
```

```python
#loading the dataset
df = pd.read_csv("D:/Programming/Datasets/non_normal_dataset.csv", squeeze=True)
print(type(df))
df.head()
```

```python
#plotting graph to see if the data is normaly distributed
sns.displot(df, kde=True)
```

```python
#Using statistics to check if the data is normally distributed
print(st.normaltest(df))

print("Dataset not normally distributed")
```

Sampling with Replacement

```python
def get_sample(data, n, seed):
    np.random.seed(seed)
    sample = np.random.choice(data, size=n, replace=True)
    return sample

data = df

test_sample = get_sample(data, 30, 0)
print(test_sample[:5])
```

Generating a Sample Mean

```python
def get_sample_mean(sample):
    sample = sample.mean()
    return sample

test_sample2 = get_sample(data, 30, 0)
test_sample2_mean = get_sample_mean(test_sample2)
print(test_sample2_mean)
```

Creating a Sampling Distribution of Sample Means

```python
def create_sample_distribution(data, dist_size=100, n=30):
    sample_means = []

    for i in range(dist_size):
        sample = get_sample(data, n, seed=i)
        sample_mean = get_sample_mean(sample)
        sample_means.append(sample_mean)

    return sample_means

data = df

test_sample_dist = create_sample_distribution(data)
print(test_sample_dist[:5]) 
```

Visualizing the Sampling Distribution as it Becomes Normal

```python
#plot of distribution size 10 and n=3
sample_dist = create_sample_distribution(data, dist_size=10, n=3)
sns.histplot(sample_dist, kde=True)
```

```python
#plot with distribution size 30 and n = 10


sample_dist2 = create_sample_distribution(data, dist_size=30, n=10)
sns.histplot(sample_dist2, kde=True)
```

```python
#plot of distribution size=1000  and n=30
sample_dist3 = create_sample_distribution(data, dist_size=1000, n=30)
sns.histplot(sample_dist3, kde=True)
```

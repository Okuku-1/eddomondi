Standard Normal Distribution

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters for the two distributions
mean1, sd1 = 5, 3  # Distribution 1
mean2, sd2 = 10, 2  # Distribution 2

# Generate random data for both distributions
d1 = np.random.normal(mean1, sd1, 1000)
d2 = np.random.normal(mean2, sd2, 1000)

# Plot the histograms and KDEs for both distributions
sns.histplot(d1, bins=30, kde=True, color='blue', label='Distribution 1')
sns.histplot(d2, bins=30, kde=True, color='orange', label='Distribution 2')

# Add titles and labels
plt.title('Histogram and KDE for Two Normal Distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```

```python
# Stardardizing and visualizing distributions

sns.histplot([(x - d1.mean())/d1.std() for x in d1]);
sns.histplot([(x - d2.mean())/d2.std() for x in d2]);
```

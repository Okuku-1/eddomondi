Normal Distribution

```python
import numpy as np

nd = np.random.normal(14, 2.8, 5000)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#ploting a histogram of normal distribution
plt.hist(nd, bins=20, density=True, color="g")
plt.xlabel("random variables")
plt.ylabel("frequency")
plt.title("Normal distribution plot");
```

```python
from scipy.stats import norm
#calculating normal distribution density
xmin, xmax = plt.xlim() 
x = np.linspace(xmin, xmax, 10)
density = norm.pdf(x, 14, 2.8)
```

```python
#plot of the histogram and density function

plt.plot(x, density, "r--")
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("Probability Density Function")
plt.show()
```

```python
import seaborn as sns
#Visualizing the distribution using seaborn and plotting the KDE
sns.histplot(nd, bins=20, kde=True, color="g", stat="density")

# Add titles and labels
plt.title('Histogram with KDE for Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("D:/Programming/Datasets/causes_of_death.tsv", delimiter= "\t")
df.head()
```

```python
#grouping by State and gender

df.groupby(["State", "Gender"]).sum()
```

```python
#Group by State, Gender, and Race and finding the average

df.groupby(["State", "Gender", "Race"]).mean()
```

```python
#Group by Gender and Race and finding the minimum value

df.groupby(["Gender", "Race"]).min()
```

```python
# Group by 'State', sum 'Deaths', sort in descending order, and plot

df.groupby("State")["Deaths"].sum().sort_values(ascending=False).plot(kind= "bar", figsize=(10,6))

#adding title and axis
plt.title("Total Number of Deaths by State")
plt.xlabel("State")
plt.ylabel("Number of Deaths")
```

Inspecting Data

```python
#Slicing the Population column
to_drop = df[df["Population"] == "Not Applicable"]

#Dropping the rows from the dataset
df.drop(to_drop.index, axis=0, inplace=True)

#Convering the Population column into integer
df["Population"] = df["Population"].astype("int64")

#Printing the result
print(df["Population"].dtype)
```

```python
#Creating a barchat of the mean population by state
df.groupby("State")["Population"].mean().sort_values(ascending=False).plot(kind="bar", figsize=(8,6))
plt.title("Mean Population by state")
plt.xlabel("States")
plt.ylabel("Mean Population")
```

```python
#Using aggregate method

grouped = df.groupby(["State", "Gender"])[["Deaths", "Population"]].agg(["mean", "min", "max", "std"]).head()
```

```python
#multi-hierarchical index
grouped.index
```

```python
# First, reset the index. Notice the subtle difference; State and Gender are now columns rather than the index
grouped = grouped.reset_index()
grouped.head()
```

```python
#Reexamining the index
grouped.index
```

```python
#Examing the columns of grouped
grouped.columns
```

```python
#value of outer layer
grouped.columns.get_level_values(0)
```

```python
#value of inner layer
grouped.columns.get_level_values(1)
```

```python
# We could also flatten these:
cols0 = grouped.columns.get_level_values(0)
cols1 = grouped.columns.get_level_values(1)
grouped.columns = [
    col0 + "_" + col1 if col1 != "" else col0 for col0, col1 in list(zip(cols0, cols1))
]
# The list comprehension above is more complicated then what we need but creates a nicer formatting and
# demonstrates using a conditional within a list comprehension.
# This simpler version works but has some tail underscores where col1 is blank:
# grouped.columns = [col0 + '_' + col1 for col0, col1 in list(zip(cols0, cols1))]
grouped.columns
```

```python
#Inspecting rows of the grouped dataframe

grouped.head()
```

Using Pivot

```python
#using pivot
pivot = grouped.pivot(index="State", columns="Gender", values="Deaths_mean")
pivot.head()
```

```python
# Again, notice the subtle difference of resetting the index:
pivot = pivot.reset_index()
pivot.head()
```

```python
#creating a bar chat of pivot data
pivot.plot(kind="barh", figsize=(15,8))
```

```python
pivot = pivot.set_index("State").plot(kind="barh", figsize=(15,8))
```

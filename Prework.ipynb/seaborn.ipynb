```python
import seaborn as sns
tips = sns.load_dataset("tips")

boxplot = sns.boxplot(data=tips["total_bill"])
```

Boxplot grouped by categories

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

More complet box plots

```python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3");
```

```python
import numpy as np
%matplotlib inline
```

```python
#Creating a sample dataset
data = np.random.normal(size=(20, 10)) + np.arange(10) / 2
```

```python
#creating a boxplot
boxplot = sns.boxplot(data=data)
```

```python
import matplotlib.pyplot as plt

sns.set_theme()
boxplot = sns.boxplot(data=data)
plt.xlabel = ("X label")
plt.ylabel = ("Y label")
plt.title("Example Boxplot")
plt.show();
```

```python
boxplot = sns.boxplot(data=data)
boxplot.set(xlabel='X Label', ylabel='Y Label', title='Example Boxplot');
```

```python
sns.boxplot(data=data).set(xlabel='X Label', ylabel='Y Label', title='Example Boxplot');
```

```python
sns.boxplot(data=data).set(xlabel='X Label', ylabel='Y Label', title='Example Boxplot')
sns.set_style("darkgrid");
```

```python
sns.boxplot(data=data).set(xlabel='X Label', ylabel='Y Label', title='Example Boxplot')
sns.set_style("darkgrid");
sns.set_context("poster")
```

```python
sns.boxplot(data=data).set(xlabel='X Label', ylabel='Y Label', title='Example Boxplot')
sns.set_style("darkgrid");
sns.set_context("notebook")
```

```python
# information on all the games played in Group 3 for the 1950 World Cup.

df.loc[(df["Year"] == 1950) & (df["Stage"] == "Group 3")]
```

```python
df.loc[(df["Year"] == 1950) & (df["Stage"] == "Group 3")]
```

```python
#Display only the Attendance column

df.loc[(df["Year"] == 1950) & (df["Stage"] == "Group 3"), ["Attendance"]]
```

```python
#length of dataset

len(df)
```

```python
# how many home games were played by the Netherlands?

len(df.loc[df["Home Team Name"] == "Netherlands"])
```

```python
#How many games were played by the Netherlands in total?

len(df.loc[(df["Home Team Name"] == "Netherlands") | (df["Away Team Name"] == "Netherlands")])
```

```python
#how many games the USA played in the 2014 World Cup.

len(df.loc[(df["Home Team Name"] == "USA") | (df["Away Team Name"] == "USA") & (df["Year"] == 2014)])
```

```python
home = len(df.loc[(df["Year"] == 2014) & (df["Home Team Name"] == "USA")])
away = len(df.loc[(df["Year"] == 2014) & (df["Away Team Name"] == "USA")])

print(home)
print(away)
```

```python
# how many countries participated in the 1986 World Cup.

wc1986 = df.loc[df["Year"] == 1986]

print(wc1986)
print(type(wc1986))
```

```python
country_number = set(wc1986.loc[:, "Home Team Name"])

print(len(country_number))
```

```python
#how many matches had 5 goals or more in total

df["Total Goals"] = df["Home Team Goals"] + df["Away Team Goals"]

more_than_five_goals = df.loc[df["Total Goals"] >= 5]

print(len(more_than_five_goals))
```
